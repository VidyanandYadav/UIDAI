```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math

# --- Step 1: Data Load Karna ---
try:
    df = pd.read_csv('final2.csv')
    print(f"Data Loaded Successfully! Total Records: {len(df)}")
except FileNotFoundError:
    print("Error: 'final2.csv' file nahi mili. Kripya check karein ki file same folder mein hai.")

# --- Step 2: Data Cleaning & Feature Engineering ---

# Date ko datetime format mein convert karna
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Total population calculate karna
df['total_population'] = df['age_0_5'] + df['age_5_17'] + df['age_18_greater']

# Division by zero se bachne ke liye 0 ko 1 se replace karna (safe side)
df['age_18_greater_safe'] = df['age_18_greater'].replace(0, 1)
df['age_5_17_safe'] = df['age_5_17'].replace(0, 1)

# Metric 1: Future School Pressure (FSP)
# Logic: Agar 0-5 wale bacche (jo abhi school nahi jate) ki sankhya school jane walo se zyada hai
# Formula: (0-5 count / 6 years) / (5-17 count / 13 years)
df['future_school_pressure'] = (df['age_0_5'] / 6) / (df['age_5_17_safe'] / 13)

# --- Step 3: Analysis Logic ---

# 1. Benford's Law (Forensic Check)
def get_first_digit(x):
    s = str(int(x))
    return int(s[0]) if s[0].isdigit() and int(s[0]) > 0 else None

df_clean = df[df['total_population'] > 0].copy()
first_digits = df_clean['total_population'].apply(get_first_digit).dropna()
digit_counts = first_digits.value_counts(normalize=True).sort_index()
benford_law = pd.Series([math.log10(1 + 1/d) for d in range(1, 10)], index=range(1, 10))

# 2. State Profiling (Demographic Quadrants)
state_profile = df.groupby('state')[['age_0_5', 'age_5_17', 'age_18_greater', 'total_population']].sum()
state_profile['pct_0_5'] = state_profile['age_0_5'] / state_profile['total_population']
state_profile['pct_18_plus'] = state_profile['age_18_greater'] / state_profile['total_population']
# Chote states ko filter karna taaki chart saaf dikhe
state_profile_filtered = state_profile[state_profile['total_population'] > 10000]

# 3. High Pressure Districts Identify karna
district_stats = df.groupby(['state', 'district'])[['age_0_5', 'age_5_17', 'age_18_greater', 'total_population']].sum()
district_stats = district_stats[district_stats['total_population'] > 1000]
district_stats['future_school_pressure'] = (district_stats['age_0_5'] / 6) / (district_stats['age_5_17'].replace(0, 1) / 13)
top_pressure_districts = district_stats.sort_values(by='future_school_pressure', ascending=False).head(10)

# 4. Momentum (Time Trend)
daily_counts = df.groupby('date')['total_population'].sum().reset_index()
daily_counts['MA_7'] = daily_counts['total_population'].rolling(window=7).mean()

# --- Step 4: Visualization (Graphs) ---
plt.figure(figsize=(20, 15))
sns.set_style("whitegrid")

# Plot 1: Benford's Law
plt.subplot(2, 2, 1)
x = np.arange(1, 10)
width = 0.35
plt.bar(x - width/2, digit_counts, width, label='Actual Data', color='royalblue')
plt.bar(x + width/2, benford_law, width, label="Benford's Law (Theoretical)", color='orange', alpha=0.7)
plt.title("Forensic Check: Data Authenticity (Benford's Law)", fontsize=14)
plt.xlabel('First Digit')
plt.ylabel('Frequency')
plt.legend()

# Plot 2: State Clusters
plt.subplot(2, 2, 2)
sns.scatterplot(data=state_profile_filtered, x='pct_18_plus', y='pct_0_5', 
                size='total_population', sizes=(100, 1000), hue='state', legend=False, palette='viridis')
plt.axhline(y=state_profile_filtered['pct_0_5'].median(), color='red', linestyle='--', alpha=0.5)
plt.axvline(x=state_profile_filtered['pct_18_plus'].median(), color='red', linestyle='--', alpha=0.5)
plt.title('State Clusters: Young Generation vs Adults', fontsize=14)
plt.xlabel('% Population 18+')
plt.ylabel('% Population 0-5')

# Plot 3: Pressure Districts
plt.subplot(2, 2, 3)
sns.barplot(x=top_pressure_districts['future_school_pressure'], y=top_pressure_districts.index.get_level_values(1), palette='Reds_r')
plt.title('Top 10 Districts: Future School Crisis Alert', fontsize=14)
plt.xlabel('Pressure Index (>1 means shortage expected)')
plt.ylabel('District')

# Plot 4: Time Trend
plt.subplot(2, 2, 4)
plt.plot(daily_counts['date'], daily_counts['total_population'], alpha=0.3, color='gray', label='Daily Count')
plt.plot(daily_counts['date'], daily_counts['MA_7'], color='green', linewidth=2, label='7-Day Average')
plt.title('Data Collection Momentum', fontsize=14)
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()

# --- Step 5: Text Summary Print Karna ---
print("\n" + "="*50)
print()
print("="*50)
print(f"\n1. Benford's Law Deviation: {(digit_counts - benford_law).abs().sum():.4f}")
print("   (Insight: Deviation suggests this is targeted campaign data, not random census.)")

print("\n2. Top Districts needing New Schools (Baby Boom Alert):")
print(top_pressure_districts[['future_school_pressure']].head(5))

print("\n3. States with Youngest Population Profile (Top 3):")
print(state_profile_filtered.sort_values(by='pct_0_5', ascending=False)[['pct_0_5']].head(3))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('final2.csv')
df['date'] = pd.to_datetime(df['date'])

# --- Unique Analysis 1: The "Missing Parents" Index (Migration Proxy) ---
# Logic: If a district has huge number of kids but very few recorded adults, 
# it implies either: 
# a) Adults are migrant workers working elsewhere (Money Order Economy)
# b) Data collection bias (only collecting kids info)
# We define "Dependency Load" = (Kids) / (Adults). 
# Normal dependency is ~0.5 to 0.8. If it's > 5.0, it's an anomaly (Ghost Workforce).

df['total_kids'] = df['age_0_5'] + df['age_5_17']
# Avoid division by zero
df['age_18_safe'] = df['age_18_greater'].replace(0, 1) 
df['migration_proxy_index'] = df['total_kids'] / df['age_18_safe']

# Group by District (filter small samples)
district_mig = df.groupby(['state', 'district'])[['total_kids', 'age_18_greater', 'migration_proxy_index']].sum()
district_mig['ratio'] = district_mig['total_kids'] / district_mig['age_18_greater'].replace(0, 1)
district_mig = district_mig[district_mig['age_18_greater'] > 100] # reliable sample

# "Labor Export Zones" (High Kids, Low Adults)
export_zones = district_mig.sort_values(by='ratio', ascending=False).head(5)
# "Labor Import Zones" (Low Kids, High Adults - likely industrial/urban hubs)
import_zones = district_mig.sort_values(by='ratio', ascending=True).head(5)


# --- Unique Analysis 2: "The Sunday Laziness" (Operational Loss) ---
# Calculate efficiency by day of the week.
df['day_name'] = df['date'].dt.day_name()
daily_performance = df.groupby('day_name')['total_kids'].mean().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
)
# Calculate "Lost Potential": If Sunday worked as hard as the best day (say Tuesday), how many more registrations?
best_day_avg = daily_performance.max()
sunday_avg = daily_performance['Sunday']
opportunity_loss_per_sunday = best_day_avg - sunday_avg


# --- Unique Analysis 3: "The Rounding Detective" (Heaping Analysis) ---
# Do surveyors count exactly or do they guess (10, 20, 50)?
# Check terminal digits of 'age_0_5'
def get_terminal_digit(x):
    return int(str(x)[-1])

df['last_digit'] = df['age_0_5'].apply(get_terminal_digit)
digit_counts = df['last_digit'].value_counts(normalize=True).sort_index()

# If data is perfect, all digits 0-9 should be ~10% (0.1).
# If 0 and 5 are spikes, humans are guessing.
rounding_bias_score = (digit_counts[0] + digit_counts[5]) - 0.2  # Excess over expected 20%


# --- Visuals ---
plt.figure(figsize=(18, 6))

# Plot 1: Migration Proxy
plt.subplot(1, 3, 1)
# Combine export and import for contrast
zones = pd.concat([export_zones, import_zones])
sns.barplot(x=zones['ratio'], y=zones.index.get_level_values(1), palette='coolwarm')
plt.title('Migration Proxy: "Labor Export" vs "Import" Districts')
plt.xlabel('Child-to-Adult Ratio (High = Parents Missing)')

# Plot 2: Day Efficiency
plt.subplot(1, 3, 2)
sns.lineplot(x=daily_performance.index, y=daily_performance.values, marker='o', color='purple')
plt.axhline(y=best_day_avg, color='green', linestyle='--', label='Max Capacity')
plt.fill_between(daily_performance.index, daily_performance.values, best_day_avg, color='red', alpha=0.1)
plt.title(f'Operational Efficiency: The Weekend Slump')
plt.ylabel('Avg Registrations per Entry')

# Plot 3: Heaping (Rounding Bias)
plt.subplot(1, 3, 3)
sns.barplot(x=digit_counts.index, y=digit_counts.values, color='teal')
plt.axhline(y=0.1, color='red', linestyle='--', label='Expected (Random)')
plt.title('Human Bias Check: Are Surveyors Guessing?')
plt.xlabel('Last Digit of Count')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.savefig('unique_analysis_v2.png')

print("Labor Export Zones (High Ratio):", export_zones['ratio'])
print("Labor Import Zones (Low Ratio):", import_zones['ratio'])
print("Rounding Bias Score (Excess 0/5s):", rounding_bias_score)
print("Opportunity Loss per Sunday (avg):", opportunity_loss_per_sunday)
```

```python
# Reach Analysis: Thane aur South 24 Parganas model districts hain (Wide coverage).

# Risk Alert: West Champaran mein coverage badhane ki zaroorat hai (Too much load on few pincodes).

# Anomaly: Panipat aur Pakur ka data highly suspicious hai (Single point dependency).





# --- Analysis 1: "Reach vs Volume" Matrix (Penetration Depth) ---
# How "deep" did the survey go? Did it cover many pincodes or just one major one?
district_reach = df.groupby(['state', 'district']).agg({
    'pincode': 'nunique',          # Breadth (How many locations)
    'age_0_5': 'sum',
    'age_5_17': 'sum',
    'age_18_greater': 'sum'
}).reset_index()

district_reach['total_pop'] = district_reach['age_0_5'] + district_reach['age_5_17'] + district_reach['age_18_greater']

# Identify Quadrants
# Median values to split the grid
median_pincodes = district_reach['pincode'].median()
median_pop = district_reach['total_pop'].median()

# "Deep Penetration" = High Pincodes, High Pop
# "Super Dense" = Low Pincodes, High Pop (Crowded spots)
deep_penetration = district_reach[
    (district_reach['pincode'] > median_pincodes) & 
    (district_reach['total_pop'] > median_pop)
].sort_values(by='total_pop', ascending=False).head(5)

super_dense = district_reach[
    (district_reach['pincode'] < median_pincodes) & 
    (district_reach['total_pop'] > median_pop)
].sort_values(by='total_pop', ascending=False).head(5)


# --- Analysis 2: Concentration Risk (Single Pincode Dependency) ---
# Which districts rely too much on just 1 or 2 pincodes?
# For each district, find the share of the top pincode.
def get_top_pincode_share(group):
    total = group['age_0_5'].sum() + group['age_5_17'].sum() + group['age_18_greater'].sum()
    if total == 0: return 0
    pincode_counts = group.groupby('pincode')[['age_0_5', 'age_5_17', 'age_18_greater']].sum().sum(axis=1)
    top_pincode_val = pincode_counts.max()
    return top_pincode_val / total

# This can be slow on large data, so let's filter for significant districts first
top_districts_list = district_reach[district_reach['total_pop'] > 5000]['district'].unique()
df_significant = df[df['district'].isin(top_districts_list)]

concentration_risk = df_significant.groupby(['state', 'district']).apply(get_top_pincode_share).sort_values(ascending=False).head(10)


# --- Analysis 3: The "Work Rhythm" Heatmap ---
# Month vs Day of Week heatmap to see operational patterns
df['month'] = df['date'].dt.month_name()
df['day_of_week'] = df['date'].dt.day_name()
# Sort order
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

heatmap_data = df.pivot_table(index='day_of_week', columns='month', values='age_0_5', aggfunc='count')
# Reindex to ensure correct order
heatmap_data = heatmap_data.reindex(days_order)
# Reindex columns (months) based on what's available
available_months = [m for m in months_order if m in heatmap_data.columns]
heatmap_data = heatmap_data[available_months]


# --- Visualization ---
plt.figure(figsize=(18, 12))

# Plot 1: Reach vs Volume
plt.subplot(2, 2, 1)
sns.scatterplot(data=district_reach, x='pincode', y='total_pop', alpha=0.6, size='total_pop', sizes=(20, 200), legend=False, color='dodgerblue')
plt.axvline(median_pincodes, color='red', linestyle='--', alpha=0.5, label='Median Reach')
plt.axhline(median_pop, color='red', linestyle='--', alpha=0.5, label='Median Vol')
# Label outliers
for i, row in deep_penetration.head(3).iterrows():
    plt.text(row['pincode'], row['total_pop'], row['district'], fontsize=9, ha='right')
plt.title('Penetration Matrix: Breadth (Pincodes) vs Volume')
plt.xlabel('Unique Pincodes Covered')
plt.ylabel('Total Population Recorded')
plt.legend()

# Plot 2: Concentration Risk
plt.subplot(2, 2, 2)
sns.barplot(x=concentration_risk.values, y=concentration_risk.index.get_level_values(1), palette='magma')
plt.title('Concentration Risk: Districts Dependent on Single Pincode')
plt.xlabel('Share of Top Pincode (1.0 = 100% data from 1 pincode)')

# Plot 3: Work Rhythm Heatmap
plt.subplot(2, 1, 2)
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=False, linewidths=.5)
plt.title('Operational Rhythm Heatmap: When is the work happening?')
plt.xlabel('Month')
plt.ylabel('Day of Week')

plt.tight_layout()
plt.savefig('advanced_analysis_v3.png')

print("Deep Penetration Districts (Good Reach):\n", deep_penetration[['district', 'pincode', 'total_pop']])
print("\nSuper Dense Districts (High Vol, Low Reach):\n", super_dense[['district', 'pincode', 'total_pop']])
print("\nHigh Concentration Risk (Top 5):\n", concentration_risk.head(5))
```

```python

# RE-CREATE CALCULATED COLUMNS
df['total_population'] = df['age_0_5'] + df['age_5_17'] + df['age_18_greater']

# --- Analysis: District Personas (Clustering without ML) ---
# We will classify districts into 4 categories based on Performance & Reliability

# 1. Aggregate Data
district_summary = df.groupby(['state', 'district']).agg({
    'total_population': 'sum',
    'pincode': 'nunique',
    'age_0_5': 'sum',
    'age_18_greater': 'sum'
}).reset_index()

# 2. Define Metrics
median_pop = district_summary['total_population'].median()
median_pincodes = district_summary['pincode'].median()

def classify_district(row):
    if row['total_population'] > median_pop and row['pincode'] > median_pincodes:
        return "Star Performer (High Vol, High Reach)"
    elif row['total_population'] > median_pop and row['pincode'] <= median_pincodes:
        return "Risk Zone (High Vol, Low Reach)"
    elif row['total_population'] <= median_pop and row['pincode'] > median_pincodes:
        return "Emerging (Low Vol, High Reach)"
    else:
        return "Needs Attention (Low Vol, Low Reach)"

district_summary['Persona'] = district_summary.apply(classify_district, axis=1)

# --- Analysis: The "Impossible" Records (Anomaly Detection) ---
# Finding specific entries that are logically impossible or highly suspicious
# Rule: > 50 kids (0-5) and 0 adults in a SINGLE entry (row)
suspicious_entries = df[(df['age_0_5'] > 50) & (df['age_18_greater'] == 0)]

# --- Visualization ---
plt.figure(figsize=(16, 8))

# Scatter Plot of Personas
plt.subplot(1, 2, 1)
# Create a temporary color map mapping
persona_counts = district_summary['Persona'].value_counts()
sns.scatterplot(
    data=district_summary, 
    x='pincode', 
    y='total_population', 
    hue='Persona', 
    style='Persona',
    palette='deep',
    s=100,
    alpha=0.7
)
plt.title('District Segmentation: Who is performing well?')
plt.xlabel('Reach (Unique Pincodes)')
plt.ylabel('Volume (Total Population)')
plt.yscale('log') # Log scale because population varies hugely
plt.legend(title='District Persona')

# Bar Plot for Suspicious Counts by State
plt.subplot(1, 2, 2)
if not suspicious_entries.empty:
    suspicious_counts = suspicious_entries['state'].value_counts().head(10)
    sns.barplot(x=suspicious_counts.values, y=suspicious_counts.index, palette='Reds_r')
    plt.title('Top States with "Impossible" Data (50+ Kids, 0 Adults)')
    plt.xlabel('Number of Suspicious Entries')
else:
    plt.text(0.5, 0.5, 'No Impossible Records Found', ha='center')

plt.tight_layout()
plt.savefig('impressive_personas.png')

# Output for narrative
print("Persona Distribution:\n", district_summary['Persona'].value_counts())
print("\nTop 5 'Star Performer' Districts:\n", district_summary[district_summary['Persona'] == "Star Performer (High Vol, High Reach)"].sort_values(by='total_population', ascending=False).head(5)[['district', 'state', 'total_population']])
print("\nMost Suspicious Entries (Ghost Orphanages?):\n", suspicious_entries[['date', 'state', 'district', 'pincode', 'age_0_5', 'age_18_greater']].head(5))
```

```python



# --- Analysis 1: Infrastructure Demand Model (Sarkar ke liye Budget Planning) ---
# Assumptions (Indian Govt Standards approx):
# 1 Anganwadi Worker per ~40-80 children (0-6 yrs). Let's take 50.
# 1 Primary School Teacher per ~30 children (RTE Act, 5-14 yrs). Let's take 30.

infra_df = df.groupby(['state', 'district'])[['age_0_5', 'age_5_17']].sum().reset_index()

# Calculate Requirements
infra_df['anganwadis_needed'] = (infra_df['age_0_5'] / 50).astype(int)
infra_df['teachers_needed'] = (infra_df['age_5_17'] / 30).astype(int)

# Top districts by requirement
top_infra_demand = infra_df.sort_values(by='anganwadis_needed', ascending=False).head(10)

# --- Analysis 2: Work Stability Index (Consistency Check) ---
# Govt wants steady work, not "month end rush".
# Calculate Coefficient of Variation (CV) of daily entries.
# CV = (Standard Deviation / Mean). Lower is better (More stable).

daily_work = df.groupby(['state', 'district', 'date'])['total_population'].sum().reset_index()
stability = daily_work.groupby(['state', 'district'])['total_population'].agg(['mean', 'std', 'count']).reset_index()

# Filter for districts with enough working days (>10 days) to be statistically significant
stability = stability[stability['count'] > 10]
stability['cv'] = stability['std'] / stability['mean']

# "Steady Eddies" (Most Consistent) vs "Panic Masters" (Least Consistent)
steady_districts = stability.sort_values(by='cv', ascending=True).head(5)
panic_districts = stability.sort_values(by='cv', ascending=False).head(5)


# --- Analysis 3: The "Survey Wave" (Geospatial Progression) ---
# Visualizing how the team moved across a state (using Pincode as proxy for location)
# We pick the state with most data: Uttar Pradesh
target_state = 'Uttar Pradesh'
wave_data = df[df['state'] == target_state].copy()

# Sort by date
wave_data = wave_data.sort_values(by='date')


# --- Visualization ---
plt.figure(figsize=(18, 12))

# Plot 1: Infrastructure Demand
plt.subplot(2, 2, 1)
# Stacked bar for Anganwadi vs Teachers
ind = np.arange(len(top_infra_demand))
width = 0.35
p1 = plt.bar(ind, top_infra_demand['anganwadis_needed'], width, label='Anganwadis Needed (0-5)', color='#ff9999')
p2 = plt.bar(ind, top_infra_demand['teachers_needed'], width, bottom=top_infra_demand['anganwadis_needed'], label='Teachers Needed (5-17)', color='#66b3ff')

plt.ylabel('Units Required')
plt.title('Infrastructure Demand Model: Budget Allocation Guide')
plt.xticks(ind, top_infra_demand['district'], rotation=45)
plt.legend()

# Plot 2: Stability Index
plt.subplot(2, 2, 2)
# Compare Top 5 Steady vs Top 5 Panic
combined_stability = pd.concat([steady_districts.assign(Type='Consistent'), panic_districts.assign(Type='Erratic')])
sns.barplot(data=combined_stability, x='cv', y='district', hue='Type', dodge=False, palette={'Consistent': 'green', 'Erratic': 'red'})
plt.title('Workforce Stability: Who works daily vs Who dumps data?')
plt.xlabel('Instability Score (CV) - Lower is Better')

# Plot 3: The "Wave" (UP)
plt.subplot(2, 1, 2)
# Scatter plot: X=Date, Y=Pincode
plt.scatter(wave_data['date'], wave_data['pincode'], alpha=0.3, s=10, c='purple')
plt.title(f'Geospatial Progression in {target_state}: Did they sweep area by area?')
plt.xlabel('Date')
plt.ylabel('Pincode (Location Proxy)')
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig('policy_planning_analysis.png')

# Output
print("INFRASTRUCTURE PLANNING (Budget Requirements):")
print(top_infra_demand[['state', 'district', 'anganwadis_needed', 'teachers_needed']])

print("\nSTABILITY AWARDS (Best Managed Teams):")
print(steady_districts[['state', 'district', 'cv']])

print("\nPANIC ZONES (Highly Erratic Work):")
print(panic_districts[['state', 'district', 'cv']])
```

```python


# --- Analysis 1: Hospital Recommendation (Where to open hospitals?) ---
# Criteria: 
# 1. High absolute count of 0-5 children (Pediatric demand).
# 2. High Density (Population / Pincodes). Areas where many people are crowded in few locations.

district_stats = df.groupby(['state', 'district']).agg({
    'age_0_5': 'sum',
    'total_population': 'sum',
    'pincode': 'nunique'
}).reset_index()

# Calculate 'Pressure Score' = (0-5 Population) * (Density Factor)
# Density Factor = Total Pop / Unique Pincodes
district_stats['density'] = district_stats['total_population'] / district_stats['pincode']
district_stats['hospital_need_index'] = district_stats['age_0_5']  # Direct demand is the best metric for government

# Top locations for Pediatric Hospitals (Child Care)
hospital_targets = district_stats.sort_values(by='hospital_need_index', ascending=False).head(10)


# --- Analysis 2: Aadhaar Awareness (Where are people 'Jagruk'?) ---
# Criteria: 
# High ratio of Adults (18+) registering themselves. 
# If only kids are registered, it's likely school/anganwadi forced. 
# If adults are registering, it's self-awareness.

district_stats['total_adults'] = df.groupby(['state', 'district'])['age_18_greater'].sum().values
district_stats['awareness_score'] = district_stats['total_adults'] / district_stats['total_population']

# Top 'Jagruk' Districts (High Adult participation)
# Filter for significant population (> 1000) to avoid skewed small samples
valid_districts = district_stats[district_stats['total_population'] > 5000]
top_aware_districts = valid_districts.sort_values(by='awareness_score', ascending=False).head(10)
# Bottom 'Awareness Needed' Districts
bottom_aware_districts = valid_districts.sort_values(by='awareness_score', ascending=True).head(10)


# --- Visualization ---
plt.figure(figsize=(18, 12))

# Plot 1: Hospital Targets
plt.subplot(2, 2, 1)
sns.barplot(x=hospital_targets['age_0_5'], y=hospital_targets['district'], palette='Reds_r')
plt.title('Priority Locations for New Pediatric Hospitals (High Child Vol)')
plt.xlabel('Number of Children (0-5 Years)')

# Plot 2: Awareness Map (Top 10)
plt.subplot(2, 2, 2)
sns.barplot(x=top_aware_districts['awareness_score'], y=top_aware_districts['district'], palette='Greens_r')
plt.title('Most "Jagruk" Districts (High Adult Participation)')
plt.xlabel('Awareness Score (% Adults)')
plt.xlim(0, 1)



print("Hospital Priority Locations (Top 5):\n", hospital_targets[['state', 'district', 'age_0_5']].head(5))
print("\nMost Aware Districts (High Adult %):\n", top_aware_districts[['state', 'district', 'awareness_score']].head(5))
print("\nLeast Aware Districts (Need Campaigns):\n", bottom_aware_districts[['state', 'district', 'awareness_score']].head(5))
```
